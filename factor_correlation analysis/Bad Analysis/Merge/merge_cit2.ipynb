```python
import pandas as pd
import os

# Set working directory
os.chdir('/Users/nsusser/Desktop/Github/happyDB/')
# Load the two matrices
matrix1_path = "dataframes/tests/gpt40-mini/updated_ratings_matrix.csv"
matrix2_path = "dataframes/tests/gpt40-mini/CIT/ratings_matrix.csv"

# Read the matrices (adjust for Excel files if needed)
matrix1 = pd.read_csv(matrix1_path, header=[0, 1], index_col=0)
matrix2 = pd.read_csv(matrix2_path, header=[0, 1], index_col=0)

# # Ensure both matrices have the same multi-index structure
# matrix1 = matrix1.reindex_like(matrix2, method=None)
# matrix2 = matrix2.reindex_like(matrix1, method=None)
print(matrix1.head())
print(matrix2.head())
```

```python
def flatten_columns_with_unique_names(matrix):
    # Ensure we construct unique names using the first two rows for metadata
    matrix.columns = [
        f"{matrix.iloc[0, idx]}_{matrix.iloc[1, idx]}_{col}"
        for idx, col in enumerate(matrix.columns)
    ]
    # Drop the metadata rows used for column names
    matrix = matrix.drop(index=matrix.index[:2]).reset_index(drop=True)
    # Rename the first column explicitly
    matrix.columns = ["cleaned_hm"] + list(matrix.columns[1:])
    return matrix

# Apply the function to both matrices
matrix1 = flatten_columns_with_unique_names(matrix1)
matrix2 = flatten_columns_with_unique_names(matrix2)


```

```python
matrix2 = matrix2.dropna(axis=1, how='all')
matrix2.head()
```

```python

merged = pd.merge(matrix1, matrix2, left_index=True, right_index=True, how='inner')
merged.rename(columns={'cleaned_hm_x': 'cleaned_hm'}, inplace=True)
merged.head()
merged.to_csv('merged.csv')

```
