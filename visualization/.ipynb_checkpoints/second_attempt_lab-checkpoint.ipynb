```python
sent_embeddings = pd.read_parquet('dataframes/embedding_parquets/sentences_with_embeddings.parquet') 
```

```python
sent_embeddings.head()
```

```python
# Convert list of embeddings into a NumPy array
sent_embeddings["embedding"] = sent_embeddings["embedding"].apply(lambda x: np.array(eval(x)) if isinstance(x, str) else np.array(x))
sentence_embeddings = np.vstack(sent_embeddings["embedding"].values)  

print("Embedding matrix shape:", sentence_embeddings.shape)  
```

```python
# Reduce 3072D sentence embeddings to 3D
reducer = umap.UMAP(n_components=3, n_neighbors=15, min_dist=0.1, random_state=42)
sentence_3d = reducer.fit_transform(sentence_embeddings) 
```

```python
# Add 3D coordinates to DataFrame
sent_embeddings["embedding_3d"] = sentence_3d.tolist()

# Save to CSV
sent_embeddings.to_csv("sentence_embeddings_3D.csv", index=False)
```

```python
gpt_results = pd.read_csv('analysis_data/results - factors.csv')
```

```python
items_path = 'dataframes/scales_clean.csv'
items = pd.read_csv(items_path)

# Clean and sanitize column names
items['Scale'] = items['Scale'].str.strip().str.replace(r"\s+", "_", regex=True)
items['Dimension'] = items['Dimension'].str.strip().str.replace(r"\s+", "_", regex=True)
items['Items'] = items['Items'].str.strip().str.replace(r"\s+", "_", regex=True)

# Create flattened column names
cols = [f"{scale}_{dimension}_{item}" for scale, dimension, item in zip(
    items['Scale'], items['Dimension'], items['Items']
)]
```

```python
#subset df with cols
df_subset = gpt_results[cols]
print(df_subset.columns)
```

```python
#standardize df_subset
scaler = StandardScaler()
df_subset = pd.DataFrame(scaler.fit_transform(df_subset), columns=df_subset.columns)
```
